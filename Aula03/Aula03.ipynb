# Exercício 6

Seu diretor comercial te solicitou a criação de alguma coisa, onde fosse possível calcular o bônus dos/as vendedores/as, de acordo com o atingimento de metas (somente com 2 casas decimais), cuja a regra é:
- Atingimento  = (Valor vendido/meta) * 100 

O Bônus é calculado da seguinte forma:
- Atingimento menor que 100: Não há bônus
- Atingimento maior ou igual a 100 e menor ou igual a 105: 2% sobre o valor vendido do/a vendedor/a
- Atingimento maior que 105: 3% sobre o valor vendido do/a vendedor/a

O salário final do/a vendedor/a deve observar a seguinte regra:
- Salário final = Salário base + bônus

Ao final, deve apresentar uma mensagem com as seguintes informações:
- Vendedores abaixo da meta:
    - Nomes:
    - Atingimento:
    - Salário final:
- Vendedores que bateram a meta entre 100 e 105:
    - Nomes:
    - Atingimento:
    - Bônus:
    - Salário final:
- Vendedores que bateram a meta acima de 105:
    - Nomes:
    - Atingimento:
    - Bônus:
    - Salário final:
    
Suponha que os dados de entrada no sistema são:

- Meta de vendas: 10.000
- Salário base dos/as vendedores/as: 2.000

- Vendedor: João
    - Valor vendido:  8.000
- Vendedora: Maria
    - Valor vendido: 10.800
- Vendedor: Sócrates
    - Valor vendido: 12.150
- Vendedora: Isis
    - Valor vendido: 12.280
- Vendedora: Ohana
    - Valor vendido: 10.390
- Vendedor: Platão
    - Valor vendido: 10.032
- Vendedora: Afrodite
    - Valor vendido: 8.950



```python
#Resposta
#criando listas
lista_nome_vendedores_abaixo = []
lista_atingimento_vendedores_abaixo = []
lista_salario_final_vendedores_abaixo = []

lista_nome_vendedores_100_105 = []
lista_atingimento_vendedores_100_105 = []
lista_bonus_vendedores_100_105 = []
lista_salario_final_vendedores_100_105 = []

lista_nome_vendedores_105 = []
lista_atingimento_vendedores_105 = []
lista_bonus_vendedores_105 = []
lista_salario_final_vendedores_105 = []
```

```python
#variáveis
meta = 10000
salario_base = 2000

#Inserindo dados
nome = input("Nome:")
valor_vendido = int(input("Valor vendido (R$):"))

#atingimento de meta
atingimento = (valor_vendido/meta)*100
atingimento = round(atingimento,2)

#cálculo do bônus
if atingimento < 100:
    lista_nome_vendedores_abaixo.append(nome)
    lista_atingimento_vendedores_abaixo.append(atingimento)
    lista_salario_final_vendedores_abaixo.append(salario_base)
elif atingimento >=100 and atingimento <=105:
    #calcula Bônus correspondente
    bonus = (valor_vendido * 2/100)
    bonus = round(bonus, 2)
    
    #calcula salário final com bônus
    salario_final = salario_base + bonus 
    salario_final = round(salario_final,2)
    
    #adiciona nas listas
    lista_nome_vendedores_100_105.append(nome)
    lista_atingimento_vendedores_100_105.append(atingimento)
    lista_bonus_vendedores_100_105.append(bonus)
    lista_salario_final_vendedores_100_105.append(salario_final)
elif atingimento > 105:
    #calcula Bônus correspondente
    bonus = (valor_vendido * 3/100) 
    bonus = round(bonus, 2)
    
    #calcula salário final com bônus
    salario_final = salario_base + bonus 
    salario_final = round(salario_final,2)
    
    #adiciona nas listas
    lista_nome_vendedores_105.append(nome)
    lista_atingimento_vendedores_105.append(atingimento)
    lista_bonus_vendedores_105.append(bonus)
    lista_salario_final_vendedores_105.append(salario_final)

print("")
print("Vendedores/as abaixo da meta:")
print("Nomes:", lista_nome_vendedores_abaixo)
print("Atingimento de meta:", lista_atingimento_vendedores_abaixo)
print("Salário final (R$):", lista_salario_final_vendedores_abaixo)

print("")
print("Vendedores/as que bateram a meta, entra 100 e 105:")
print("Nomes:", lista_nome_vendedores_100_105)
print("Atingimento de meta:", lista_atingimento_vendedores_100_105)
print("Bônus (R$):", lista_bonus_vendedores_100_105)
print("Salário final (R$):", lista_salario_final_vendedores_100_105)

print("")
print("Vendedores/as que bateram a meta, acima de 105:")
print("Nomes:", lista_nome_vendedores_105)
print("Atingimento de meta:", lista_atingimento_vendedores_105)
print("Bônus (R$):", lista_bonus_vendedores_105)
print("Salário final (R$):", lista_salario_final_vendedores_105)
```

# ESTRUTURA DE REPETIÇÃO

As estruturas de repetiçao são popularmente conhecidas como "loop", pois executam a mesma instrução pela quantidade de vezes em que for instruída. Elas também são chamada de "laços de repetição".

Como já sabemos, o Python utiliza-se da identação para execução de blocos de código! A regra segue firme e forte! :)

Em python, temos 2 funções para trabalharmos com estruturas de repetição:

- for: Nos possibilita percorrer em todos os itens de um determinado conjunto de dados.
- while: Nos possibilita percorrer em um conjunto de dados, de acordo com uma condição específica. Isso quuer dizer que, enquanto a condição é TRUE, o script segue executando a ação.

```python
#exemplo FOR - i retornando o dado
for i in lista_nome_vendedores_105:
    print (i)
```

```python
#exemplo FOR / ELSE
for i in lista_nome_vendedores_105:
    print (i)
else:
    print("Todos os vendedores com atingimento superior a 105 foram exibidos!")
```

```python
#exemplo FOR / ELSE / FOR
print("Vendedores com atingimento superior a 105:")
for i in lista_nome_vendedores_105:
    print (i)
else:
    print("")
    print("Vendedores que não atingiram a meta")
    for i in lista_nome_vendedores_abaixo:
        print (i)
        
```

```python
#Exemplo While
i = 0 #iniciar variável condicional de loop
while i < 5: #a função while requer que a variável condicional do loop, nesse caso o "i", seja iniciada antes de "entrar" no loop
    print(i)
    #i += 1
    i = i + 1
```

Exemplo FOR/WHILE

O Diretor comercial percebeu que alguns vendedores superaram muito a meta e, por conta disso, resolveu acrescentar um bônus para todos/as que atingiram 120 ou mais.

O cálculo desse bônus será o seguinte:

A cada 1 ponto acima de 120, acrescentar R$ 150.

(atingimento de meta - 120)*150

Dessa forma, ele te solicitou esse ajuste e pediu para que no final fosse exibida uma mensagem com o nome, o bônus de excelência e o salário final, acrescido do bônus.

```python
#variáveis
atingimento_excelencia = 120
valor_bonus_excelencia = 150
lista_vendedores_excelencia = []
lista_salario_excelencia = []
lista_bonus_excelencia = []

#O índice utilizado para controlar o loop, em geral representado por "i" (porém, pode ser qualquer outra letra ou nome) NÃO suporta números decimais
#para esses casos, utilizar a função len e range, para que o "i" se torne um contador do tipo inteiro.
#Um outro cenário que será necessário utilizar len e range é quando há a necessidade de se acessar índices (0,1,2,3...n) em listas.
for i in range(len(lista_atingimento_vendedores_105)):
    if (lista_atingimento_vendedores_105[i] >= atingimento_excelencia):
        #Cálculo do bônus
        bonus_excelencia = (lista_atingimento_vendedores_105[i] - atingimento_excelencia)*valor_bonus_excelencia
        bonus_excelencia = round(bonus_excelencia,2)
                
        #Cálculo do salário
        salario_excelencia = lista_salario_final_vendedores_105[i] + bonus_excelencia
        salario_excelencia = round(salario_excelencia,2)
        
        #adiciona as listas
        lista_vendedores_excelencia.append(lista_nome_vendedores_105[i])
        lista_bonus_excelencia.append(bonus_excelencia)
        lista_salario_excelencia.append(salario_excelencia)
else:
    impressao_vendedores = input("Bônus de excelência calculado. Deseja visualizar os dados? (S para sim ou N para não)")
    if (impressao_vendedores.upper() == "S"):
        i = 0
        print("")
        print("Vendedores de excelência:")
        print("*************************")
        while i < len(lista_vendedores_excelencia):
            print("- Nome:", lista_vendedores_excelencia[i])
            print("- Bônus:", lista_bonus_excelencia[i])
            print("- Salário excelência:", lista_salario_excelencia[i])
            print("")
            #i = i + 1
            i+=1
    else:
        print("Aplicação finalizada. Obrigado.")
```
