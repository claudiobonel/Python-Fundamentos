Colete os dados do brasileirão de 2020 e faça o seguinte:
- Uma análise descritiva (baseado no que já aprendemos) em algum dos valores
- Comparação entre os anos de 2020, 2021 e 2022, de algum valor
- Exporte o resultado para o Excel

Link para coleta dos dados: https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol_de_2020_-_S%C3%A9rie_A

```python
#coleta tabela do brasileiro de 2021
import pandas as pd

#macth = "Classificação". Retorna a tabela próxima ao texto.
#[0] é o único índice da lista.
tabela2020 = pd.DataFrame(pd.read_html("https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol_de_2020_-_S%C3%A9rie_A",match="Classificação")[0])

#lista_tabelas.append(tabela2021)

#exibir tabela
tabela2020
```

```python
#biblioteca com métodos
#numpy = Numerical Python. Biblioteca para cálculos númericos
import numpy as np
from statistics import mode
from pathlib import Path
lista_V = []
#Analisando vitórias
   
media = np.average(tabela2020["V"])
dados = ("Média",media)
lista_V.append(dados)

mediana = np.median(tabela2020["V"])
dados = ("Mediana",mediana)
lista_V.append(dados)

#moda em statistcs
moda = mode(tabela2020["V"])
dados = ("Moda",moda)
lista_V.append(dados)

quantile_25 = np.quantile(tabela2020["V"],q=0.25)
dados = ("Quantil 25%",quantile_25)
lista_V.append(dados)

quantile_50 = np.quantile(tabela2020["V"],q=0.5)
dados = ("Quantil 50%",quantile_50)
lista_V.append(dados)

quantile_75 = np.quantile(tabela2020["V"],q=0.75)
dados = ("Quantil 75%",quantile_75)
lista_V.append(dados)

amplitude = np.max(tabela2020["V"]) - np.min(tabela2020["V"])
dados = ("Amplitude",amplitude)
lista_V.append(dados)

variancia = round(np.var(tabela2020["V"]),2)
dados = ("Variância",variancia)
lista_V.append(dados)

#Média de vitórias 2020 - 2021
lista_comparacoes = []
media_vitorias2020 = (np.sum(tabela2020["V"])/np.sum(tabela2020["J"]))
dados = ("Média de vitórias 2020:",media_vitorias2020)
lista_comparacoes.append(dados)

media_vitorias2021 = (np.sum(tabela2021["V"])/np.sum(tabela2020["J"]))
dados = ("Média de vitórias 2021:",media_vitorias2021)
lista_comparacoes.append(dados)

media_vitorias2022 = (np.sum(tabela2022["V"])/np.sum(tabela2020["J"]))
dados = ("Média de vitórias 2022:",media_vitorias2022)
lista_comparacoes.append(dados)

print("2020 - Análise Descritiva de vitórias")
print("*************************************")
print("Média:", media) 
print("Mediana:", mediana)
print("Moda:", moda)
print("Quantil 25%:",quantile_25)
print("Quantil 50%:",quantile_50)
print("Quantil 75%:",quantile_75)
print("Amplitude:", amplitude)
print("Variancia:", variancia)  
print("")
print("Comparação da média de vitórias entre os anos de 2020,2021,2022")
print("***************************************************************")
print("Média de vitórias 2020:",media_vitorias2020)
print("Média de vitórias 2021:",media_vitorias2021)
print("Média de vitórias 2022:",media_vitorias2022)

df_descritiva = pd.DataFrame(lista_V)
df_descritiva = df_descritiva.rename(columns={0:"Análise",1:"Valor"})

df_comparacoes = pd.DataFrame(lista_comparacoes)
df_comparacoes = df_comparacoes.rename(columns={0:"Comparação",1:"Valor"})

#definindo endereço no sistema de arquivos
endereco = Path("C:\\Users\\claud\\OneDrive\\Claudio Bonel-DADOTECA\\Senac\\Python - FATEC\\Dados\\")

#nome do arquivo
excel_v_por_jogo = endereco / "v_jogo_2021_2022.xlsx"

if (endereco.exists()):
    #Se o endereço existir segue com o processo de criação ou append de dados no excel
    escrita = pd.ExcelWriter(excel_v_por_jogo, engine='xlsxwriter')
    
    #Adiciona na planilha "Descritiva"
    df_descritiva.to_excel(escrita,sheet_name="Descritiva", index=False)
    
    #adiciona na planilha "Comparacoes"
    df_comparacoes.to_excel(escrita,sheet_name="Comparacoes", index=False)
    
    #fecha o arquivo
    escrita.close()
    
else:
    #Se o endereço não existir, envia uma mensagem de erro!
    print('Endereço não existe! Favor verificar!')
    
print("Arquivo exportado com sucesso!")
```
