```python
#biblioteca com métodos
#numpy = Numerical Python. Biblioteca para cálculos númericos
import numpy as np
from statistics import mode
from pathlib import Path
lista_V = []
#Analisando vitórias
   
media = np.average(tabela2020["V"])
dados = ("Média",media)
lista_V.append(dados)

mediana = np.median(tabela2020["V"])
dados = ("Mediana",mediana)
lista_V.append(dados)

#moda em statistcs
moda = mode(tabela2020["V"])
dados = ("Moda",moda)
lista_V.append(dados)

quantile_25 = np.quantile(tabela2020["V"],q=0.25)
dados = ("Quantil 25%",quantile_25)
lista_V.append(dados)

quantile_50 = np.quantile(tabela2020["V"],q=0.5)
dados = ("Quantil 50%",quantile_50)
lista_V.append(dados)

quantile_75 = np.quantile(tabela2020["V"],q=0.75)
dados = ("Quantil 75%",quantile_75)
lista_V.append(dados)

amplitude = np.max(tabela2020["V"]) - np.min(tabela2020["V"])
dados = ("Amplitude",amplitude)
lista_V.append(dados)

variancia = round(np.var(tabela2020["V"]),2)
dados = ("Variância",variancia)
lista_V.append(dados)

#Média de vitórias 2020 - 2021
lista_comparacoes = []
media_vitorias2020 = (np.sum(tabela2020["V"])/np.sum(tabela2020["J"]))
dados = ("Média de vitórias 2020:",media_vitorias2020)
lista_comparacoes.append(dados)

media_vitorias2021 = (np.sum(tabela2021["V"])/np.sum(tabela2020["J"]))
dados = ("Média de vitórias 2021:",media_vitorias2021)
lista_comparacoes.append(dados)

media_vitorias2022 = (np.sum(tabela2022["V"])/np.sum(tabela2020["J"]))
dados = ("Média de vitórias 2022:",media_vitorias2022)
lista_comparacoes.append(dados)

print("2020 - Análise Descritiva de vitórias")
print("*************************************")
print("Média:", media) 
print("Mediana:", mediana)
print("Moda:", moda)
print("Quantil 25%:",quantile_25)
print("Quantil 50%:",quantile_50)
print("Quantil 75%:",quantile_75)
print("Amplitude:", amplitude)
print("Variancia:", variancia)  
print("")
print("Comparação da média de vitórias entre os anos de 2020,2021,2022")
print("***************************************************************")
print("Média de vitórias 2020:",media_vitorias2020)
print("Média de vitórias 2021:",media_vitorias2021)
print("Média de vitórias 2022:",media_vitorias2022)

df_descritiva = pd.DataFrame(lista_V)
df_descritiva = df_descritiva.rename(columns={0:"Análise",1:"Valor"})

df_comparacoes = pd.DataFrame(lista_comparacoes)
df_comparacoes = df_comparacoes.rename(columns={0:"Comparação",1:"Valor"})

#definindo endereço no sistema de arquivos
endereco = Path("C:\\Users\\claud\\OneDrive\\Claudio Bonel-DADOTECA\\Senac\\Python - FATEC\\Dados\\")

#nome do arquivo
excel_v_por_jogo = endereco / "v_jogo_2021_2022.xlsx"

if (endereco.exists()):
    #Se o endereço existir segue com o processo de criação ou append de dados no excel
    escrita = pd.ExcelWriter(excel_v_por_jogo, engine='xlsxwriter')
    
    #Adiciona na planilha "Descritiva"
    df_descritiva.to_excel(escrita,sheet_name="Descritiva", index=False)
    
    #adiciona na planilha "Comparacoes"
    df_comparacoes.to_excel(escrita,sheet_name="Comparacoes", index=False)
    
    #fecha o arquivo
    escrita.close()
    
else:
    #Se o endereço não existir, envia uma mensagem de erro!
    print('Endereço não existe! Favor verificar!')
    
print("Arquivo exportado com sucesso!")
```

# Coletando dados de CSV

```python
#Coleta de ocorrências
ocorrencias = pd.read_csv("C:\\Users\\claud\\OneDrive\\Claudio Bonel-DADOTECA\\Senac\\Python - FATEC\\Dados\\ocorrencias.csv")

#exibir DF
ocorrencias
```

# Relacionamento de DataFrames

Para DataFrames de colunas e linhas distintas, porém com uma coluna de chave de relacionamento (ou seja, de mesmo conteúdo em ambos), deve-se utilizar um método que permita a junção pela cardinalidade, no nosso caso utilizaremos o merge. É importante que as colunas de relacionamento possuam o mesmo nome, certifique-se de renomea-las.

```python
#relaciona dados do Dataframe de ocorrências com o de municípios
#O objetivo é recuperar o nome de cada município, através do cod mcirc (município da circunscrição)
#DataFrames de uma relação cardinal 1 para N, com linhas e colunas distintas
#df_ocorrencias = ocorrencias.set_index('mcirc').join(municipio.set_index('mcirc'))
df_ocorrencias = pd.merge(ocorrencias,municipio,on=["mcirc","mcirc"],how="inner")

#visualiza o resultado do relacionamento
#df_ocorrencias

#Realiza um filtro de ano, para trazer somente os anos acima de 2021
#filtro com número
df_ocorrencias_maior_igual_ano = df_ocorrencias.query("ano >= 2021")
#filtro com texto
df_ocorrencias_maior_igual_ano = df_ocorrencias_maior_igual_ano.query("ocorrencias == 'roubo_veiculo' or ocorrencias == 'furto_veiculos'")

#visualiza o resultado do relacionamento com filtro
df_ocorrencias_maior_igual_ano
```

```python
#seleciona somente as colunas de interesse
df_ocorrencias_maior_igual_ano = df_ocorrencias_maior_igual_ano[["municipios","qtde"]]

df_ocorrencias_maior_igual_ano
```

# Ranqueando dados

```python
#agrupa valores por categoria
df_municipios_rj = pd.DataFrame(df_ocorrencias_maior_igual_ano.groupby(["municipios"])["qtde"].sum()) #.sort_values(by="qtde",ascending=True)

#Reseta o index, para que munícipio seja um dado
df_municipios_rj = df_municipios_rj.reset_index()

#Criando um ranking top N
df_municipios_rj = df_municipios_rj.nlargest(10,"qtde")

df_municipios_rj
```

```python
#Calculando o fator de influência de cada cidade sobre o top 10

#Total ocorrências - Top 10
total_ocorrencias = np.sum(df_municipios_rj["qtde"])

#Cálculo do fator de influência
df_municipios_rj["influencia"] = round((df_municipios_rj["qtde"]/total_ocorrencias),4)

#Ordenando o dataframe para o ranqking
df_municipios_rj = df_municipios_rj.sort_values(by="influencia",ascending=False)

df_municipios_rj
```

# Visualizando dados

```python
#biblioteca para visualização de dados
import matplotlib.pyplot as plt

#tamanho do gráfico
plt.figure(figsize=(13, 5))

#É preciso ordenar o dataframe pela métrica que deseja ranquear o gráfico.
#Barras horizontais a ordenação precisa ser inversa a que vc deseja exibir no gráfico.
df_municipios_rj_gr = df_municipios_rj.sort_values(by="influencia",ascending=True)

#plotagem do gráfico
rank_municipios = plt.barh("municipios",width="influencia", height=0.6,data=df_municipios_rj_gr)

#título do gráfico
plt.title("Ranking de municípios") 

#remove o eixo X
ax = plt.gca() #get current axis: método para acionar um eixo específico
ax.get_xaxis().set_visible(False) #remove o eixo X

#rótulos de dados
plt.bar_label(rank_municipios, padding=10)

#removendo borda
plt.box(False) 

#exportação da imagem
endereco_rk_municipios_jpg = endereco / 'rk_municipios.png'
plt.savefig(endereco_rk_municipios_jpg,pad_inches=1)

#Retornando gráfico
plt.show(rank_municipios)
```

```python
#nome do arquivo
excel_rk_municipios = endereco / "rk_municipios.xlsx"

if (endereco.exists()):
    #Se o endereço existir segue com o processo de criação ou append de dados no excel
    escrita = pd.ExcelWriter(excel_rk_municipios, engine='xlsxwriter')
    
    #Adiciona na planilha "Tabela"
    df_municipios_rj.to_excel(escrita,sheet_name="Ranking", index=False)
    
    #adiciona na planilha "Gráfico"
    planilha = escrita.sheets["Ranking"]
    planilha.insert_image('E2', endereco_rk_municipios_jpg)
  
    #fecha o arquivo
    escrita.close()
    
else:
    #Se o endereço não existir, envia uma mensagem de erro!
    print('Endereço não existe! Favor verificar!')
    
print("Arquivo exportado com sucesso!")
```
